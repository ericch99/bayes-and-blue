```
from PIL import Image  
import numpy as np
import os
os.chdir("/content/drive/My Drive/ProjectX 2020/Intelligent_Downsampling")
  
# Opens a image in RGB mode  
A_img = np.load(r"../Data/1979/2/CycleGAN_Data/ExtremeWeather/1979_0_2.npy")
A_img = (A_img - np.min(A_img))/(np.max(A_img)-np.min(A_img))
A_img = np.transpose(A_img, (2, 0, 1))[0];
im = Image.fromarray(np.uint8(A_img * 255) , 'L')
  

im = im.resize((144, 96), resample=Image.BICUBIC) 
# Shows the image in image viewer  
im.save("results/1979_0_2.png")  
```
